```python
import cv2
img = cv2.imread('./images/lena.jpg')#shortcut
cv2.imshow("original",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Lossless
Saving the image in PNG format without losing quality.

```python
cv2.imwrite('./images/Lena_Lossless.png', img, [cv2.IMWRITE_PNG_COMPRESSION, 0])
#IMWRITE_PNG_COMPRESSION must be in the (0, 9) interval—the bigger the number, 
#the smaller the file on the disk, but the slower the decoding process.
```

### Lossy
Saving the image in the JPEG format.

```python
cv2.imwrite('./images/Lena_Lossy.jpg', img, [cv2.IMWRITE_JPEG_QUALITY, 0])
#When saving to JPEG format, we can manage the compression process by setting the value of IMWRITE_JPEG_QUALITY.
#We can set this as any value from 0 to 100. But here, we have a situation where bigger is better. 
#Larger values lead to higher result quality and a lower amount of JPEG artifacts.
```
